```python
# load cleaned data

import pandas as pd

X = pd.read_csv("X_clean.csv")
y = pd.read_csv("y_clean.csv").squeeze("columns")

print("X shape:", X.shape)
X.describe()
```

```python
D1 = X.copy()

print("D1 (raw) — min/max per feature:")
D1.agg(["min", "max"]).T
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
D2 = pd.DataFrame(
    minmax_scaler.fit_transform(D1),
    columns=D1.columns
)

print("D2 (min-max scaled) — min/max per feature:")
D2.agg(["min", "max"]).T
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
D3 = pd.DataFrame(
    standard_scaler.fit_transform(D1),
    columns=D1.columns
)

print("D3 (standard scaled) — mean/std per feature:")
D3.agg(["mean", "std"]).T
```

```python
comparison = pd.DataFrame({
    "Total (D1)": D1["Total"],
    "Total (D2)": D2["Total"],
    "Total (D3)": D3["Total"],
    "Speed (D1)": D1["Speed"],
    "Speed (D2)": D2["Speed"],
    "Speed (D3)": D3["Speed"],
})
comparison.head()
```

```python
D1.to_csv("D1.csv", index=False)
D2.to_csv("D2.csv", index=False)
D3.to_csv("D3.csv", index=False)
y.to_csv("y_clean.csv", index=False)

print("Saved D1.csv, D2.csv, D3.csv")
print(f"Each has shape: {D1.shape}")
```
