```python
# load cleaned data

import pandas as pd

X = pd.read_csv("X_clean.csv")
# squeeze("columns") converts the single-column y dataframe into a pandas Series
y = pd.read_csv("y_clean.csv").squeeze("columns")

print("X shape:", X.shape)
# describe() gives count/mean/std/min/max per column
X.describe()
```

```python
# Define D1 as the original, unscaled dataset (the baseline for comparison)
D1 = X.copy()

# Show min/max per feature to confirm the raw stats are on very different scales
print("D1 (raw) — min/max per feature:")
D1.agg(["min", "max"]).T
```

```python
#  Create D2 by applying Min-Max scaling to D1
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler rescales each feature independently so its values fall strictly within [0, 1]
minmax_scaler = MinMaxScaler()
D2 = pd.DataFrame(
    minmax_scaler.fit_transform(D1),
    columns=D1.columns
)

# Confirm scaling worked: every feature's min should now be 0 and max should be 1
print("D2 (min-max scaled) — min/max per feature:")
D2.agg(["min", "max"]).T
```

```python
# Create D3 by applying Standard scaling (z-score normalization) to D1
from sklearn.preprocessing import StandardScaler

# StandardScaler transforms each feature to have mean 0 and standard deviation 1
# (subtracts the mean, then divides by the standard deviation), unlike MinMax which bounds to [0,1]
standard_scaler = StandardScaler()
D3 = pd.DataFrame(
    standard_scaler.fit_transform(D1),
    columns=D1.columns
)

# Confirm scaling worked. every feature's mean should now be ~0 and std should be ~1
print("D3 (standard scaled) — mean/std per feature:")
D3.agg(["mean", "std"]).T
```

```python
# Side-by-side comparison of two features across all three scaling versions
# Lets us visually confirm D1, D2, D3 represent the same underlying values on different scales
comparison = pd.DataFrame({
    "Total (D1)": D1["Total"],
    "Total (D2)": D2["Total"],
    "Total (D3)": D3["Total"],
    "Speed (D1)": D1["Speed"],
    "Speed (D2)": D2["Speed"],
    "Speed (D3)": D3["Speed"],
})
comparison.head()
```

```python
# Save D1, D2, D3 (and the label) to folder so classification/clustering notebooks can load them
D1.to_csv("D1.csv", index=False)
D2.to_csv("D2.csv", index=False)
D3.to_csv("D3.csv", index=False)

# Re-save y here too so all downstream files (features + label) live together
y.to_csv("y_clean.csv", index=False)

print("Saved D1.csv, D2.csv, D3.csv")
# All three datasets should have identical shape (only the scaling differs, not the row/column counts)
print(f"Each has shape: {D1.shape}")
```
