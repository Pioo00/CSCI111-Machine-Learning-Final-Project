```python
import pandas as pd 

# Loading and printing the 
df = pd.read_csv("Pokemon.csv")
print("Shape:", df.shape)
df.head()
```

```python
# Handling duplicates
# The ID column just repeats for all the duplicates of the pokemon
# Use Name instead to check for duplicates
# Keep only the first occurrence of each Pokédex number
df_clean = df.drop_duplicates(subset="#", keep="first")

# Save the cleaned dataset
df_clean.to_csv("cleaned_pokemon.csv", index=False)

print(f"Original rows: {len(df)}")
print(f"Cleaned rows: {len(df_clean)}")
print("Cleaned dataset saved as 'cleaned_pokemon.csv'")


```

```python
# Extracting the 7 numeric columns that will be used as model inputs 

feature_cols = ["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
X = df_clean[feature_cols].copy()

print("X shape:", X.shape)
X.head()
```

```python
# Extracting Legendary column and converting it to numeric (1 = Legendary, 0 = not)

y = df_clean["Legendary"].astype(int)

print()
```

```python
# Check for missing values in X or y if there are any 

print("Missing values in X:")
print(X.isna().sum())
print("\nMissing values in y:", y.isna().sum())
```

```python
# Saving cleaned data
X.to_csv("X_clean.csv", index=False)
y.to_csv("y_clean.csv", index=False)

print("Saved X_clean.csv and y_clean.csv")
print(f"Final dataset: {X.shape[0]} rows, {X.shape[1]} features, {y.sum()} Legendary")
```
